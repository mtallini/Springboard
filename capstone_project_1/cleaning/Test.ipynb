```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
directory = "/home/michael/Desktop/capstone_project_1/data"
sample = pd.read_csv(directory + "/sample.csv")
stopwords = set(stopwords.words('english'))
ps = PorterStemmer()
tfidf = TfidfVectorizer()
```

```python
def row_tokenize(row):
    words = word_tokenize(row)
    vocab = [ps.stem(word) for word in words if (word not in stopwords)]
    return ' '.join(vocab)
```

```python
sample['tokenized']= sample['review/text'].str.lower().apply(row_tokenize)
sample.head()
```

```python
tfidf.fit(sample['tokenized'])
X = tfidf.transform(sample['tokenized'])
```

```python
sample.dropna(inplace=True)
sample['liked?'] = np.where(sample['review/score'] > 3, 1, 0)
sample.head()
```

```python
X = sample['review/text']
y = sample['liked?']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print("Train set has total {0} entries with {1:.2f}% negative, {2:.2f}% positive".format(len(X_train),
                                                                             (len(X_train[y_train == 0]) / (len(X_train)*1.))*100,
                                                                        (len(X_train[y_train == 1]) / (len(X_train)*1.))*100))
print("Test set has total {0} entries with {1:.2f}% negative, {2:.2f}% positive".format(len(X_test),
                                                                             (len(X_test[y_test == 0]) / (len(X_test)*1.))*100,
                                                                            (len(X_test[y_test == 1]) / (len(X_test)*1.))*100))
```
